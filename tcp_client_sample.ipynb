## connect and send module

```python
import socket
ACK_DATA = '\x02' +  'ACK0' +  '\x03'
HOST = 'localhost'  # 서버 IP 주소
PORT = 8006  # 서버 포트 번호

client_socket = socket.socket(socket.AF_INET, socket.SOCK_STREAM)  # 소켓 생성
client_socket.connect((HOST, PORT))  # 서버에 연결
```

```python
# 서버로부터 응답 수신
response = client_socket.recv(4096)
print(f"서버로부터 받은 응답: {response.decode('euc-kr')}")
client_socket.sendall(ACK_DATA.encode('euc-kr'))
```
